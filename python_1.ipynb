```python
#or we can do this, indeed we supposed to
lokasi = pd.read_csv("Lokasi (1).csv")
lokasi.head()
```

```python
#buat peta dasar
peta = folium.Map(location=[1.50422578928316, 101.940465010793], zoom_start=13)
```

```python
#fungsi warna
def warna(nama_lokasi):
    colors = {
        'Kantor desa api-api': 'red',
        'SDN 1 Bandar Laksamana': 'orange',
        'SMPN 1 Bandar Laksamana': 'green',
        'Masjid Al-Falah': 'pink',
        'Masjid Syuhada': 'purple',
        'Masjid An-Nur': 'darkred',
        'Kantor BPD' : 'beige',
        'Musholla Darul Naim':'forestgreen',
        'Perumahan Nelayan Camar Laut': 'darkblue',
        'Posyandu': 'cadetblue',
        'Pantai Indah Api Api': 'blue',
        'Perkebunan Warga':'gray'
    }
    return colors.get(nama_lokasi)
```

```python
#fungsi untuk semua titik
for idx, row in lokasi.iterrows():
    #fungsi text pop-up
    popup_text = f"Lokasi: {row['Nama']}"
    #masukkan warna
    color = warna(row['Nama'])
    
    marker = Marker(
        [row['Latitude'], row['Longitude']], 
        popup=popup_text,
        icon=folium.Icon(color=color)
    )
    marker.add_to(peta)
```
